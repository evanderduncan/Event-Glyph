```python
import cv2
import numpy as np
from pathlib import Path

# ── config ──────────────────────────
image_path = Path("sourcebook/pages/page_054.png")
output_dir = Path("sourcebook/dataset")
output_dir.mkdir(parents=True, exist_ok=True)

COLS         = 6
ROWS         = 7
SYMBOL_RATIO = 0.72
margin       = 15

# ── load ─────────────────────────────
img  = cv2.imread(str(image_path))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(f"Loaded: {img.shape}")

# ── compute grid ─────────────────────
h, w  = img.shape[:2]
cell_w = (w - 2 * margin) // COLS
cell_h = (h - 2 * margin) // ROWS
print(f"Cell size: {cell_w} x {cell_h}")

# ── draw grid preview ─────────────────
preview = img.copy()

for row in range(ROWS):
    for col in range(COLS):
        x1 = margin + col * cell_w
        y1 = margin + row * cell_h
        x2 = x1 + cell_w
        y2 = y1 + cell_h
        split_y = y1 + int(cell_h * SYMBOL_RATIO)

        # green = cell border, orange = symbol/label split
        cv2.rectangle(preview, (x1, y1), (x2, y2), (0, 180, 90), 2)
        cv2.line(preview, (x1, split_y), (x2, split_y), (0, 140, 255), 2)

# ── save ─────────────────────────────
out_path = output_dir / "page_054_preview.png"
result   = cv2.imwrite(str(out_path), preview)
print(f"imwrite returned: {result}")
print(f"Preview saved to: {out_path.resolve()}")
print(f"File exists: {out_path.exists()}")
```

```python
import cv2
import numpy as np
from pathlib import Path

def find_all_grid_lines(gray, line_width_threshold=0.55, min_gap=50):
    h, w = gray.shape
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    inv = 255 - binary

    lines = []
    last_line_y = -min_gap

    for y in range(0, h):
        row_sum = inv[y, :].sum()
        if row_sum > (line_width_threshold * w * 255):
            if y - last_line_y > min_gap:
                lines.append(y)
                last_line_y = y

    return lines


def find_grid_extent(gray):
    lines = find_all_grid_lines(gray)

    if len(lines) < 2:
        print("  Warning: fewer than 2 grid lines found, using fallback margins")
        h = gray.shape[0]
        return 210, h - 410

    grid_top    = lines[0]
    grid_bottom = lines[-1]

    print(f"  Found {len(lines)} border lines")
    print(f"  Grid extent: y={grid_top} to y={grid_bottom}")

    return grid_top, grid_bottom
```

```python
import cv2
from pathlib import Path

# Test on a range of known awkward pages
test_pages = [
    "page_054.png",  # clean full grid
    "page_038.png",  # likely has overflow from previous section
    "page_080.png",  # likely has intro paragraph at bottom
    "page_120.png",  # another section transition
]

for page in test_pages:
    img  = cv2.imread(f"sourcebook/pages/{page}")
    if img is None:
        print(f"{page} — not found, skipping")
        continue
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    lines = find_all_grid_lines(gray)
    top, bottom = find_grid_extent(gray)
    print(f"\n{page}")
    print(f"  All lines at: {lines}")
    print(f"  Using grid: {top} → {bottom}")

    # Save a preview for each
    preview = img.copy()
    h, w = img.shape[:2]
    cv2.line(preview, (0, top),    (w, top),    (0, 200, 80),  3)
    cv2.line(preview, (0, bottom), (w, bottom), (0, 100, 255), 3)
    cv2.imwrite(f"sourcebook/dataset/{page.replace('.png','')}_extent.png", preview)
    print(f"  Preview saved")
```

```python
def detect_grid_lines(gray, min_line_length_ratio=0.3):
    """
    Use Hough line detection to find the actual printed grid lines.
    Returns sorted lists of x positions (vertical lines) and 
    y positions (horizontal lines).
    """
    h, w = gray.shape

    # Edge detection first
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)

    # Detect lines
    lines = cv2.HoughLinesP(
        edges,
        rho=1,
        theta=np.pi/180,
        threshold=100,
        minLineLength=int(min(w, h) * min_line_length_ratio),
        maxLineGap=20
    )

    if lines is None:
        print("  No lines detected — falling back to margin estimates")
        return None, None

    h_lines = []  # horizontal — these give row boundaries
    v_lines = []  # vertical   — these give col boundaries

    for line in lines:
        x1, y1, x2, y2 = line[0]
        angle = abs(np.degrees(np.arctan2(y2 - y1, x2 - x1)))

        # Near-horizontal lines (within 5 degrees)
        if angle < 5 or angle > 175:
            y_mid = (y1 + y2) // 2
            h_lines.append(y_mid)

        # Near-vertical lines (within 5 degrees of 90)
        elif 85 < angle < 95:
            x_mid = (x1 + x2) // 2
            v_lines.append(x_mid)

    # Cluster nearby lines — many detections per physical line
    def cluster(positions, gap=20):
        if not positions:
            return []
        positions = sorted(positions)
        clusters = [[positions[0]]]
        for p in positions[1:]:
            if p - clusters[-1][-1] < gap:
                clusters[-1].append(p)
            else:
                clusters.append([p])
        return [int(np.mean(c)) for c in clusters]

    h_clustered = cluster(h_lines)
    v_clustered = cluster(v_lines)

    print(f"  Detected {len(h_clustered)} horizontal lines: {h_clustered}")
    print(f"  Detected {len(v_clustered)} vertical lines:   {v_clustered}")

    return h_clustered, v_clustered


def get_cell_boundaries_auto(img):
    """
    Build cell list directly from detected grid lines.
    Falls back to margin-based approach if detection fails.
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    h, w = img.shape[:2]

    h_lines, v_lines = detect_grid_lines(gray)

    # Need at least 2 of each to form cells
    if not h_lines or not v_lines or len(h_lines) < 2 or len(v_lines) < 2:
        print("  Falling back to fixed margins")
        return get_cell_boundaries(img)  # your margin-based fallback

    cells = []
    for row_idx in range(len(h_lines) - 1):
        for col_idx in range(len(v_lines) - 1):
            x1 = v_lines[col_idx]
            y1 = h_lines[row_idx]
            x2 = v_lines[col_idx + 1]
            y2 = h_lines[row_idx + 1]
            cells.append((x1, y1, x2, y2, row_idx, col_idx))

    print(f"  Built {len(cells)} cells from detected lines")
    return cells
```

```python
def get_cell_boundaries(img, cols=6, rows=7, left_margin=70, right_margin=80):
    """Margin-based fallback if line detection fails."""
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    h, w = img.shape[:2]

    grid_top, grid_bottom = find_grid_extent(gray)

    usable_w = w - left_margin - right_margin
    usable_h = grid_bottom - grid_top
    cell_w   = usable_w // cols
    cell_h   = usable_h // rows

    cells = []
    for row in range(rows):
        for col in range(cols):
            x1 = left_margin + col * cell_w
            y1 = grid_top    + row * cell_h
            x2 = x1 + cell_w
            y2 = y1 + cell_h
            cells.append((x1, y1, x2, y2, row, col))

    return cells
```
